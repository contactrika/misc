```python
%reset -f
%pylab inline
import GPy
import GPyOpt
from numpy.random import seed
```

```python
# A function to optimize (a shifted quadratic; used for illustrative purposes).
def myfunction(x):
    return cos(2*x)+cos(5*x)+cos(30*x) + 0.1*rand()
```

```python
bounds = [{'name': 'var_1', 'type': 'continuous', 'domain': (-1,1)}]
seed(2)
myProblem = GPyOpt.methods.BayesianOptimization(myfunction,domain=bounds,
                                                initial_design_numdata=3,normalize_Y=False,
                                                model_type= 'GP_MCMC', acquisition_type='EI_MCMC')
myProblem.plot_acquisition()
print 'x_opt', myProblem.x_opt
print 'fx_opt', myProblem.fx_opt
print myProblem.model.model
```

```python
myProblem.run_optimization(max_iter=1)
myProblem.plot_acquisition()
print 'x_opt', myProblem.x_opt
print 'fx_opt', myProblem.fx_opt
print myProblem.model.model
```
